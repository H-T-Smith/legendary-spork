**Task 1: Create a version of Wordle that takes a word at random from a dataset of words. Add a basic UI that allows for either a user or a computer to make guesses.**

```python
import numpy as np


infile = open('/content/drive/MyDrive/valid-wordle-words.txt','r')

words = infile.readlines()

# Remove the newline character from each word
for i in range(len(words)):
  words[i]=words[i][0:5]

def wordle_playable():
  print("Welcome to wordle! Type a 5 letter word to begin. Note all characters are lowercase. Type 'quit' to quit at any time. \n")
  print("Guide:\n")
  print("     0: letter not in word, gets removed from t\n")
  print ("    1: letter in word in the incorrect spot")
  print("     2: letter in word and in the correct spot.\n")


  answer=np.random.choice(words)


  alphabet='abcdefghijklmnopqrstuvwxyz'
  tries=0
  guess=' '

  while not(guess == answer):
    guess=input()
    if (guess=='quit'):
      return

    elif guess in words:
      guess_info=""
      for pos in range(5):
        if guess[pos] == answer[pos]:
          guess_info+='2'
        elif guess[pos] in answer:
          guess_info+='1'
        else:
          guess_info+='0'
          alphabet = alphabet.replace(guess[pos],'_')
      print(guess, "   ", guess_info, "              ", alphabet)
      print('\n')
      tries+=1

    else:
      print("\nNot in word list")

  print("You guessed the right word in",tries,"tries.")

play_again=True
while play_again:

  print("To play wordle, enter 'play'. Type anything else to quit")
  command=input()
  if not(command == 'play'):
    play_again=False
  if play_again:
    wordle_playable()


```
